```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function1(x):
    return 0.5* np.sin(2*x)+0.2

x = np.linspace(0,4, 100)
plt.xlabel('t')
plt.ylabel('x(t)')
ax = plt.gca()
ax.set_ylim([-2, 2])
ax.set_xlim([0,4])
#plt.fill_between(x, function1(x))
plt.plot(x, function1(x))
plt.show()
```

```python
def function1_squared(x):
    return (0.5* np.sin(2*x)+0.2)**2

x = np.linspace(0,4, 100)
plt.xlabel('t')
plt.ylabel('x(t)')
ax = plt.gca()
ax.set_ylim([-2, 2])
ax.set_xlim([0,4])
plt.fill_between(x, function1_squared(x))
plt.plot(x, function1_squared(x))
plt.show()
```

```python
def function1_squared(x):
    return (0.5* np.sin(2*x)+0.2)**2 

x = np.linspace(0,4, 100)
plt.xlabel('t')
plt.ylabel('x(t)')
ax = plt.gca()
ax.set_ylim([-2, 2])
ax.set_xlim([0,4])
weights = np.linspace(3, 3, len(x))
plt.fill_between(x, function1_squared(x) * weights)
plt.plot(x, function1_squared(x) * weights)
plt.show()
```
